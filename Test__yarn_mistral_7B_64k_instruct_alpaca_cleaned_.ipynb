### Choice 1 - Paste the PDF **URL** address.
1. Find a PDF online **without login any account.**
2. Copy the PDF url link like this sample: https://web.mit.edu/watsan/Docs/Other%20Documents/KAF/Pandey%20-%20Effects%20of%20Air%20Space%20in%20ABF%202004.pdf
3. Right click the file to copy the path.
4. Replace the url address in **"url"** variable.
5. Change the range of pages you want to test.

```python
import io
import requests
from pypdf import PdfReader

# Replace the PDF address you want to test.(Cannot replace the PDF url before you need to login something)
url = 'https://web.mit.edu/watsan/Docs/Other%20Documents/KAF/Pandey%20-%20Effects%20of%20Air%20Space%20in%20ABF%202004.pdf'

r = requests.get(url)
f = io.BytesIO(r.content)

reader = PdfReader(f)

# Change the rage of page you want to test.
Starting_Page = 2
Ending_Page = 4

Text_PDF_C1 = ""
for page_num in range(Starting_Page-1, Ending_Page, 1):
    page = reader.pages[page_num]
    Text_PDF_C1 += page.extract_text()
print(len(Text_PDF_C1))
```
